```python
#Se cargan las librerias
import pandas as pd
import numpy as np
import plotly. express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df=pd.read_csv('/content/Limpieza_Base_Spotify.csv')
df
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
#Contar el modo de la canción
gender_counts = df['Modo'].value_counts()

# Gráfico de pastel
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90,
         colors=['#A405F2', '#F205F2'])
plt.title('Modo')
plt.show()
```

```python
df['Año_Lanzamiento'] = pd.to_numeric(df['Año_Lanzamiento'], errors='coerce').astype('Int64')
df['Año_Lanzamiento'].unique()
```

```python
df['Mes_Lanzamiento'] = pd.to_numeric(df['Mes_Lanzamiento'], errors='coerce').astype('Int64')
df['Mes_Lanzamiento'].unique()
```

```python
df['Día_Lanzamiento'] = pd.to_numeric(df['Día_Lanzamiento'], errors='coerce').astype('Int64')
df['Día_Lanzamiento'].unique()
```

```python
df['Participantes'] = pd.to_numeric(df['Participantes'], errors='coerce').astype('Int64')
df['Participantes'].unique()
```

```python
#se grafica de forma interactiva con plotly el boxplot (se puede usar seaborn o matplotlib)
fig = px.box(df, y='Participantes', title='Detección de datos atípicos')
fig.update_layout(width=300,height=500)
fig.show()
```

```python
#Se define el cuartil 75 y se le resta el cuartil 25
iqr = df['Participantes'].quantile(0.75) - df['Participantes'].quantile(0.25)
#Desarrollamos filtros superior o inferior por lo general es el iqr por 1.5 pero se puede manejar el número
filtro_inferior = df['Participantes'] > df['Participantes'].quantile(0.25) - (iqr * 1.5)
filtro_superior = df['Participantes'] < df['Participantes'].quantile(0.75) + (iqr * 1.3)

df_filtrado = df[filtro_inferior & filtro_superior]

#Graficando el boxplot
fig = px.box(df_filtrado, y='Participantes', title='Datos atípicos eliminados')
fig.update_layout(width=300,height=500)
fig.show()
```

```python
# o con seaborn
# Generar el boxplot con Seaborn
plt.figure(figsize=(3, 5))  # Ajusta el tamaño de la figura
sns.boxplot(y='Participantes', data=df_filtrado)

# Agregar título y etiquetas
plt.title('Detección de datos atípicos')
plt.ylabel('Participantes')
plt.show()
```

```python
# Crear figura y subplots (2 filas, 2 columnas)
fig, axes = plt.subplots(2,2, figsize=(12, 10))
# KDE para Año de Lanzamiento
sns.kdeplot(df['Año_Lanzamiento'], fill=True, ax=axes[0, 0])
axes[0, 0].set_title('Año de Lanzamiento')
# KDE para Mes de Lanzamiento
sns.kdeplot(df['Mes_Lanzamiento'], fill=True, ax=axes[0, 1])
axes[0, 1].set_title('Mes de Lanzamiento')
# KDE para Día de Lazamiento
sns.kdeplot(df['Día_Lanzamiento'], fill=True, ax=axes[1, 0])
axes[1, 0].set_title('Día de Lanzamiento')
# KDE para Participantes
sns.kdeplot(df['Participantes'], fill=True, ax=axes[1, 1])
axes[1, 1].set_title('Número de Participantes')
# Ajustar el layout para evitar superposición de títulos y etiquetas
plt.tight_layout()
# Mostrar gráfico
plt.show()
```
