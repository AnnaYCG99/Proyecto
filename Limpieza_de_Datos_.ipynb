```python
#Renombramiento de las columnas
df=df.rename(columns={'track_name': 'Nombre_Canción','artist(s)_name': 'Artista(s)','artist_count':'Participantes','released_year':'Año_Lanzamiento',
                      'released_month':'Mes_Lanzamiento', 'released_day':'Día_Lanzamiento','in_spotify_playlists':'Listas_Spotify',
                      'in_spotify_charts':'Porpularidad_Spotify','streams':'Reproducciones_Spotify','in_apple_playlists':'Listas_Apple',
                      'in_apple_charts':'Popularidad_Apple','in_deezer_playlists':'Listas_Deezer','in_deezer_charts':'Popularidad_Deezer',
                      'in_shazam_charts':'Popularidad_Shazam','bpm':'Bpm','key':'Tonalidad','mode':'Modo','danceability_%':'Bailabilidad_%',
                      'valence_%':'Positividad_%','energy_%':'Energía_%','acousticness_%':'Acústica_%','instrumentalness_%':'Instrumentalización_%',
                      'liveness_%':'EnVivo_%','speechiness_%':'Palabras_%','cover_url':'Cover_url'})
df
```

```python
#Lista de columnas a int
int_col=['Participantes', 'Año_Lanzamiento', 'Mes_Lanzamiento',
       'Día_Lanzamiento','Reproducciones_Spotify','Listas_Spotify', 'Porpularidad_Spotify', 'Listas_Apple',
       'Popularidad_Apple', 'Listas_Deezer', 'Popularidad_Deezer', 'Popularidad_Shazam',
       'Bpm','Bailabilidad_%','Positividad_%','Energía_%','Acústica_%',
       'Instrumentalización_%','EnVivo_%','Palabras_%']

for i in int_col:
   df[i] = pd.to_numeric(df[i], errors='coerce')
   df[i] = df[i][df[i].notna()].astype(float)
   df[i] = df[i][df[i].notna()].astype(int)

df.info()
```

```python
#Eliminamos los datos duplicados de la columna Nombre_Canción
df=df.drop_duplicates(subset=['Nombre_Canción'])
df
```

```python
#Comprobamos que ya no haya valores Duplicados en la columna Nombre_Canción
df.duplicated(subset=['Nombre_Canción']).sum()
```

```python
#Remplazamos "aaaaa", por valores Nulos en df
df=df.replace("aaaaa", np.nan, regex=True)
df
```

```python
df.isnull().sum()
```

```python
#Eliminamos los valores NaN de las siguentes columnas
df=df.dropna(subset=['Nombre_Canción','Artista(s)','Tonalidad','Modo','Cover_url'])
df
```

```python
df.isnull().sum()
```

```python
# De las siguientes columnas, reemplzamos los valores NaN, por el promedio,
# así rellenamos la base de datos, con datos númericos, para un mejor manejo de la base.
df['Listas_Spotify'].fillna(df['Listas_Spotify'].median(), inplace=True)
df['Porpularidad_Spotify'].fillna(df['Porpularidad_Spotify'].median(), inplace=True)
df['Listas_Apple'].fillna(df['Listas_Apple'].median(), inplace=True)
df['Popularidad_Apple'].fillna(df['Popularidad_Apple'].median(), inplace=True)
df['Listas_Deezer'].fillna(df['Listas_Deezer'].median(), inplace=True)
df['Popularidad_Deezer'].fillna(df['Popularidad_Deezer'].median(), inplace=True)
df['Popularidad_Shazam'].fillna(df['Popularidad_Shazam'].median(), inplace=True)
df['Bpm'].fillna(df['Bpm'].median(), inplace=True)
df['Bailabilidad_%'].fillna(df['Bailabilidad_%'].median(), inplace=True)
df['Positividad_%'].fillna(df['Positividad_%'].median(), inplace=True)
df['Energía_%'].fillna(df['Energía_%'].median(), inplace=True)
df['Acústica_%'].fillna(df['Acústica_%'].median(), inplace=True)
df['Instrumentalización_%'].fillna(df['Instrumentalización_%'].median(), inplace=True)
df['EnVivo_%'].fillna(df['EnVivo_%'].median(), inplace=True)
df['Palabras_%'].fillna(df['Palabras_%'].median(), inplace=True)
df
```

```python
# De las siguientes filas, los NaN son reemplzados por "0"
df['Año_Lanzamiento'].fillna(0, inplace=True)
df['Mes_Lanzamiento'].fillna(0, inplace=True)
df['Día_Lanzamiento'].fillna(0, inplace=True)
df['Participantes'].fillna(0, inplace=True)
df['Reproducciones_Spotify'].fillna(0, inplace=True)
df
```
